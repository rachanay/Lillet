```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
reviewlist = []
i=1
for i in range (1, 4):
    URL= 'https://www.amazon.fr/LILLET-19831/product-reviews/B007T29DA0/ref=cm_cr_arp_d_viewopt_srt?sortBy=recent'
    URL= URL + '&pageNumber=' + str(i)
    headers= {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36"}
    Page= requests.get(URL, headers= headers)
    Soup1 = BeautifulSoup(Page.content, "html.parser")
    Soup2 = BeautifulSoup(Soup1.prettify(),"html.parser")
    reviews = Soup2.findAll('div', {'data-hook':"review"})
    for item in reviews:
                    
                    review = {
                        'product' : Soup2.find('a', {'data-hook':"product-link"}).text,
                        'name' : item.find('span', {'class':"a-profile-name"}).text,
                        'rating': item.find('i', {'data-hook':"review-star-rating"}).text,
            'ratingtitle' : item.find('a', {'data-hook':"review-title"}).text, 
            'date' : item.find('span', {'data-hook':"review-date"}).text,
            'format' : item.find('a', {'data-hook':"format-strip"}).text,
            'comment' : item.find('span', {'data-hook':"review-body"}).text,
          
            
                    }
                    reviewlist.append(review)
                    i=i+1
#print(reviewlist)
i=5
for i in range (5, 50):
    URL= 'https://www.amazon.fr/LILLET-19831/product-reviews/B007T29DA0/ref=cm_cr_arp_d_viewopt_srt?sortBy=recent'
    URL= URL + '&pageNumber=' + str(i)
    headers= {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36"}
    Page= requests.get(URL, headers= headers)
    Soup1 = BeautifulSoup(Page.content, "html.parser")
    Soup2 = BeautifulSoup(Soup1.prettify(),"html.parser")
    reviews = Soup2.findAll('div', {'data-hook':"review"})
    for item in reviews:
                    
                    review = {
                        'product' : Soup2.find('a', {'data-hook':"product-link"}).text,
                        'name' : item.find('span', {'class':"a-profile-name"}).text,
                        'rating': item.find('i', {'data-hook':"cmps-review-star-rating"}).text,
            'ratingtitle' : item.find('span', {'data-hook':"review-title"}).text, 
            'date' : item.find('span', {'data-hook':"review-date"}).text,
            'format' : item.find('a', {'data-hook':"format-strip"}).text,
            'comment' : item.find('span', {'data-hook':"review-body"}).text,
          
            
                    }
                    reviewlist.append(review)
                    i=i+1
df = pd.DataFrame(reviewlist)
df.to_excel('lillet.xlsx', index=False)
```
